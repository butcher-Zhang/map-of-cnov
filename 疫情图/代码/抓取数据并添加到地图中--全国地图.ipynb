```python
url = 'http://ncovdata.spbeen.com/apis/get_china_provinces/?query_date=2020-02-20'
```

```python
import requests
```

```python
response = requests.get(url)
```

```python
response.text
```

```python
data = response.json()
```

```python
data = data['data']
```

```python
data
```

```python
province_name =[]
confirmed_all =[]
suspected_all =[]
cured_all = []
dead_all = []
for d in data:
    k = "".join(d.keys())
    cscd =d[k]
    try:
        confirmed = cscd['confirmed']
        suspected = cscd['suspected']
        cured = cscd['cured']
        dead = cscd['dead']
    except:
        if k == '青海':
            confirmed = 18
            suspected = 0
            cured = 18
            dead = 0
        else:
            confirmed = 1
            suspected = 0
            cured = 1
            dead = 0
    print(k,confirmed,suspected,cured,dead)
    confirmed_all.append(confirmed)
    suspected_all.append(suspected)
    cured_all.append(cured)
    dead_all.append(dead)
    province_name.append(k)
```

```python
from pyecharts import Map
```

```python
china = Map("中国地图",'中国疫情地图【2020-02-20】',width=800,height=600)
```

```python
china.add("确诊病例", province_name , confirmed_all , maptype='china',visual_range=[0,max(confirmed_all)//20] , is_visualmap=True , isule_text_color="#000", is_label_show=True)
china.add("疑似病例", province_name , suspected_all , maptype='china',visual_range=[0,max(confirmed_all)//20] , is_visualmap=True , isule_text_color="#000", is_label_show=True)
china.add("治愈病例", province_name , cured_all , maptype='china',visual_range=[0,max(confirmed_all)//20] , is_visualmap=True , isule_text_color="#000", is_label_show=True)
china.add("死亡病例", province_name , dead_all , maptype='china',visual_range=[0,max(confirmed_all)//20] , is_visualmap=True , isule_text_color="#000", is_label_show=True)
```

```python
china.render(path="province_map/"+"中国疫情地图_map.html")
```
