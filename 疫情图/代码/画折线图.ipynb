```python
qz = [1,4,18,9,8,13]
ys = [3,6,23,8,5,6]
```

```python
line = Line('疫情变化图')
```

```python
x_line = ['2-1','2-2','2-3','2-4','2-5','2-6']
line.add('新增确诊',x_line,qz,is_label_show=True)#对应点的名称，x轴坐标名，数据，是否展示表
```
